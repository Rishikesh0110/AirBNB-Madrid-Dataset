```python
#Checking skewness of columns 	latitude,price,minimum_nghts,Number_of_reviews,calculated_host-countings,& Availability_365
```

```python
print(air_df["price"].skew())
print(air_df["latitude"].skew())
print(air_df["minimum_nights"].skew())
print(air_df["number_of_reviews"].skew())
print(air_df["calculated_host_listings_count"].skew())
print(air_df["availability_365"].skew())
```

Observations
1. Column price, minimum_nights,number_of_reviews & calculated_host_listings_count are skewed in nature we will apply min-max scaller on these columns
2. Columns latitude & availability_365  as skeweness value lies b/w (-.5 to +.5)are skewes soo we will apply standard scaler

```python
#importing scikit-learn --user 
from sklearn.preprocessing import StandardScaler ,MinMaxScaler
```

```python
# Standardize 'latitude' and "availability_365' columns 
scaler = StandardScaler()
air_df[['latitude', 'availability_365']] = scaler.fit_transform(air_df[['latitude', 'availability_365']])
```

```python
## Standardize ' price", "minimum_nights","number_of_reviews" & "calculated_host_listings_count"' columns 
minmaxscaler = MinMaxScaler() 
air_df[["price","minimum_nights","number_of_reviews","calculated_host_listings_count"]] = minmaxscaler.fit_transform(air_df[["price","minimum_nights","number_of_reviews","calculated_host_listings_count"]])
```
